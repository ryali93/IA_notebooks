```python
import numpy as np
import scipy as sp

import matplotlib.pyplot as plt
```

## Descenso de gradiente:

$sin(\frac{1}{2} x^{2} - \frac{1}{4} y^{2} + 3) cos(2x + 1 - e^{y}])$

<img src="https://upload.wikimedia.org/wikipedia/commons/6/68/Gradient_ascent_%28surface%29.png" width="400" height="400" />

```python
func = lambda th: np.sin(1 / 2 * th[0] ** 2 - 1 / 4 * th[1] ** 2 + 3) * np.cos(2 * th[0] + 1 - np.e ** th[1])

res = 100

_X = np.linspace(-2, 2, res)
_Y = np.linspace(-2, 2, res)

_Z = np.zeros((res,res))

for ix, x in enumerate(_X):
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func([x, y])

plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()
plt.show()
```

Se agrega un punto aleatorio al cual calcular su gradiente

```python
Theta = np.random.rand(2) * 4 -2

plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()
plt.plot(Theta[0], Theta[1], "o", c="red")
plt.show()
```

```python
_T = np.copy(Theta)

h = 0.001
lr = 0.001         # lr: ratio de aprendizaje

plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()
plt.plot(Theta[0], Theta[1], "o", c="white")

grad = np.zeros(2)

for _ in range(10000):
    for it, th in enumerate(Theta):
        _T = np.copy(Theta)
        _T[it] = _T[it] + h
        deriv = (func(_T) - func(Theta)) / h
        grad[it] = deriv

    Theta = Theta - lr * grad
   
#     print(func(Theta))
    if(_ % 100 == 0):
        plt.plot(Theta[0], Theta[1], ".", c="red")

plt.plot(Theta[0], Theta[1], "o", c="green")
plt.show()
```
